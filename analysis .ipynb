```python
import pandas as pd
df = pd.read_csv("D:Edu/Faculty/Courses/Sections/cloud computing/assingmnent 2/books.csv")
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.drop(columns=["isbn","isbn13","original_publication_year","image_url","small_image_url","language_code","book_id","goodreads_book_id","best_book_id","work_id","books_count"], inplace=True)
df
```

```python
harry_potter_df = df[df['original_title'].str.contains("Harry Potter", case=False)]
harry_potter_df
```

```python
most_selling_books = harry_potter_df.sort_values(by="ratings_count", ascending=False)

print(most_selling_books[["title", "authors", "ratings_count"]].head().to_string(index=False))
```

```python
average_rating = harry_potter_df["average_rating"].mean()

print("Average rating of Harry Potter books:", average_rating)
```
